```python
from sklearn.linear_model import LinearRegression
import numpy as np

hw_train = np.array([[168, 85], [175, 66], [183, 95], [162, 52], [188, 92], [178, 63], [165, 58], [195, 85]])
bmi = hw_train[:,1] / (hw_train[:,0]/100) ** 2

hw_test = np.array([[170, 60], [185, 95]])

model = LinearRegression()
model.fit(hw_train, bmi)

print('기울기 : ', model.coef_)
print('절편 : ', model.intercept_)
print('\n')

y_LinR = model.predict(hw_test)

print('테스트 데이터 1의 BMI 값 : {} 으로 정상'.format(y_LinR[0]))
print('테스트 데이터 2의 BMI 값 : {} 으로 정상'.format(y_LinR[1]))
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

diabetes = load_diabetes()

x = diabetes.data[:, 2]
y = diabetes.target

linear_regression = LinearRegression()
x2 = x.reshape(-1, 1)
linear_regression.fit(x2, y)

plt.scatter(x2, y, alpha=0.5)
plt.plot(x2, linear_regression.predict(x2), color='red', linewidth=2)
plt.xlabel("BMI")
plt.ylabel("progression of diabetes")
plt.title("The example of Linear Regression")
plt.show()
```
