```python
# numpy 임포트하고 버전 확인하기
# 배열과 관련된 표준 라이브러리
# 다차원 배열를 다룰 수 있게 해줌
# as -> alias : 약어
import numpy as np
print(np.__version__)
```

```python
# 배열 만들기 : array()
my_arr = np.array([[10, 20, 30], [40, 50, 60]])
print(my_arr)
```

```python
# type() 함수로 객체의 타입 확인하기
# print() 함수를 사용하지 않아도 코드 셀의 마지막 줄은 자동으로 결과를 출력해줌
type(my_arr)
```

```python
# 배열에서 요소 선택하기
my_arr[0][2]
```

```python
# 넘파이 내장 함수 사용하기
np.sum(my_arr) # 모든 데이터를 더해줌
```

```python
# matplotlib
# 그래프와 관련된 표준 라이브러리
import matplotlib.pyplot as plt
# 그래프 그리기
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) # plot(x, y) x축과 y축의 값을 파이썬 리스트로 전달함
plt.show()
```

```python
# 산점도 그리기 - scatter(x, y)
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

```python
# 평균이 0 이고, 표준편차가 1인 정규분포에서 난수를 발생시킴 (랜덤한 수)
# 표준 정규 분포를 따르는 난수 1,000개를 만듦
# 정규분포에서 뽑았기에 (0, 0) 사이에 많이 분포되어 있음
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```
